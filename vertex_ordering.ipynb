### 1. User Defined UTILITY FUNCTIONS

```python
#Function Name :- to_char
#Arguments :- list of integer type
#Returns :- list of character type
#Sample to_char([1,2,3,4,5]) returns ['1','2','3','4','5']
def to_char(my_list):
    count=0
    for x in my_list :
        temp=chr(x+48)
        my_list[count]=temp
        count+=1
    return my_list

#Function Name :- check_equality
#Arguments :- two lists
#Returns :- True if both lists are equal . Else False 
#Sample check_equality([1,2,3,4,5],[4,5,6,7,8]) returns False
def check_equality(arr1,arr2):
    count=0
    end1=len(arr1)
    end2=len(arr2)
    if end1==end2 :
        while count<end1:
            if arr1[count]==arr2[count] :
                ok=''
            else :
                return False
            count+=1
    else :
        return False
    return True

#Function Name :- append
#Arguments :- three lists and an integer
#Returns :- list1 with list2 appended
#Sample append([1,2,3,2,3,1],[3,2,1],0,[1,2,3]) return [1,2,3,2,3,1,3,2,1]
def append(created,current_ordering,counter,total_nodes):
    num_nodes=len(total_nodes)
    start=num_nodes*counter
    count=0
    while count<num_nodes:
        created[start+count]=current_ordering[count]
        count+=1
    return created  

#Function Name :- sort_list
#Arguments :- list
#Returns :- sorted list 
#Sample sort_list([4,1,3,2]) returns [1,2,3,4]
def sort_list(queue):
    n=len(queue)
    for x in range(n):
        for idx in range(0, n-x-1):
            if queue[idx].cost>queue[idx+1].cost:
                temp = queue[idx]
                queue[idx] = queue[idx+1]
                queue[idx+1] = temp
    return queue

#Function Name :- check_existence
#Arguments :- two lists and an integer
#Returns :- True if both list1 exists in list2 . Else False 
#Sample check_existence([1,2,3,2,3,1],[3,2,1],0) return False
def check_existence(test,created,count):
    start=0
    end=len(test)
    iterate=0
    while iterate < count :
        my_list=created[start:end]
        test=list(test)
        if check_equality(my_list,test) :
            return True
        start=end
        end+=len(test)
        iterate+=1
    return False
```

### 4. Implementing SEARCH Algorithms

```python
#Function Name :- bfs
#Arguments :- root node
#Returns :- node with minimum cost
#Description :- Breadth First Search traverses all nodes of the tree width wise and at the end returns the one with minimum cost
def bfs(node):
    queue=[]
    if node is None:
        return
    queue.append(node)
    min_node=node
    while len(queue) > 0:
        cur_node = queue.pop(0)

        if cur_node.cost<min_node.cost:
            min_node=cur_node
        if cur_node.left is not None:
            queue.append(cur_node.left)

        if cur_node.right is not None:
            queue.append(cur_node.right)
    return min_node   

#Function Name :- ucs
#Arguments :- root node
#Returns :- node with minimum cost
#Description :- Uniform Cost Search traverses all nodes of the tree cost-wise and at the end returns the one with minimum cost
def ucs(node):
    queue=[]
    if node is None:
        return
    queue.append(node)
    min_node=node
    while len(queue) > 0:
        queue = sort_list(queue)
        cur_node = queue.pop(0)

        if cur_node.cost<min_node.cost:
            min_node=cur_node
        if cur_node.left is not None:
            queue.append(cur_node.left)

        if cur_node.right is not None:
            queue.append(cur_node.right)
    return min_node

#Function Name :- dfs
#Arguments :- root node
#Returns :- node with minimum cost
#Description :- Depth First Search traverses all nodes of the tree depth-wise and at the end returns the one with minimum cost
def dfs(node,visited,min_node):
    if node not in visited and node is not None: 
        visited.append(node)
        temp_cost=node.cost
        if temp_cost<min_node.cost :
            min_node=node
        left=dfs(node.left,visited,min_node)
        if left and left.cost<min_node.cost:
            min_node=left
        right=dfs(node.right,visited,min_node)
        if right and right.cost<min_node.cost:
            min_node=right
    return min_node
```

### 5. Graph Class and its Member Functions

```python
# Node Struct constitute a single node the tree . It contains a particular ordering along with its cost and id
class Node:
    def __init__(self, ordering,cost,node_id):
        self.left = None
        self.right = None
        self.ordering = ordering
        self.cost= cost
        self.node_id=node_id
        
# Graph CLass contains all the functionality to build a graph along with its search techniques      
class Graph:
    def __init__(self):
        self.root=None
        self.counter=0
        self.count=0
        
#Function Name :- build
#Arguments :- root node
#Returns :- graph 
#Description :- Starting from a single permuatation , it creates all the possible permutations alongside creating a graph
    def build(self,h,created,total_nodes,num_permutations):
        
        if self.counter>=num_permutations:
            return None
        
        my_ordering=np.random.permutation(total_nodes)
        current_ordering=my_ordering.tolist()

        while check_existence(my_ordering,created,self.counter) :
            my_ordering=np.random.permutation(total_nodes)
            current_ordering=my_ordering.tolist()
    
        created=append(created,current_ordering,self.counter,total_nodes)
        if self.counter==0:
            self.root=Node(current_ordering,total_ordering_min_cost(current_ordering),self.counter+1)
            node=self.root
            self.counter+=1
        elif self.counter<num_permutations :
            node=Node(current_ordering,total_ordering_min_cost(current_ordering),self.counter+1)
            self.counter+=1
        else :
            node=None
            
        if total_ordering_min_cost(current_ordering)>490 :
            self.count+=1
        
        if h == 1:
            return node
                    
        if node :
            node.left = self.build(h-1,created,total_nodes,num_permutations)
            node.right =self.build(h-1,created,total_nodes,num_permutations)
        return node
    
#Function Name :- getLevel
#Arguments :- node
#Returns :- level of a ceratin node 
#Description :- Starting from a root , it returns height of the node passed as argument
    def getLevel(self,node):
        return getLevelUtil(self.root, node, 1) 
    
    
#All 3 Search Algorithms (BFS,DFS,UCS) employing indirect recursion mechanism    
    def DFS_traversal(self):
        node=self.root
        visited=[]
        return dfs(node,visited,self.root)
     
    def BFS_traversal(self):
        node=self.root
        return bfs(node) 
    
    def UCS_traversal(self):
        node=self.root
        return ucs(node) 
 
 #Function Name :- getLevelUtil
#Arguments :- node
#Returns :- level of a ceratin node 
#Description :- Starting from a root , it returns height of the node passed as argument
def getLevelUtil(node, desired_node, level):
    if (node == None):
        return 0
    if (node == desired_node):
        return level
    
    downlevel = getLevelUtil(node.left,desired_node, level + 1)
    if (downlevel != 0):
        return downlevel

    downlevel = getLevelUtil(node.right,desired_node, level + 1)
    return downlevel 

# cALLING All 3 Search Algorithms (BFS,DFS,UCS) along with DISPLAYING results according to problem requirement   
def DFS_data(my_graph):
    print()
    print("...........DFS...........")
    print()
    min_node=my_graph.DFS_traversal()
    print("The minimum cost is for ordering : ",min_node.ordering)
    print("The COST for that ordering is : ",min_node.cost)
    print("The level of that node in the tree is : ",my_graph.getLevel(min_node))
    
def BFS_data(my_graph):
    print()
    print("...........BFS...........")
    print()
    min_node=my_graph.BFS_traversal()
    print("The minimum cost is for ordering : ",min_node.ordering)
    print("The COST for that ordering is : ",min_node.cost)
    print("The level of that node in the tree is : ",my_graph.getLevel(min_node))    
    
def UCS_data(my_graph):
    print()
    print("...........UCS...........")
    print()
    min_node=my_graph.UCS_traversal()
    print("The minimum cost is for ordering : ",min_node.ordering)
    print("The COST for that ordering is : ",min_node.cost)
    print("The level of that node in the tree is : ",my_graph.getLevel(min_node))
```
